```python
import pandas as pd

def calculate_accuracy(data, data2):
    """
    Compares the actual and generated mode/category columns and calculates accuracy.

    Parameters:
    data (pandas.DataFrame): The actual DataFrame.
    data2 (pandas.DataFrame): The generated DataFrame.

    Returns:
    dict: A dictionary containing mode and category accuracies.
    """

    # Ensure both dataframes have the same number of rows
    if len(data) != len(data2):
        raise ValueError("The actual and generated DataFrames must have the same number of rows")

    # Check if Mode and Category match between actual and generated data
    mode_matches = data['Mode'] == data2['Mode']
    category_matches = data['Category'] == data2['Category']

    # Calculate accuracy for Mode and Category
    mode_accuracy = mode_matches.sum() / len(data) * 100
    category_accuracy = category_matches.sum() / len(data) * 100
    overall_accuracy = (mode_matches & category_matches).sum() / len(data) * 100

    return {
        'Mode Accuracy': mode_accuracy,
        'Category Accuracy': category_accuracy,
        'Overall Accuracy': overall_accuracy
    }

# Assuming data contains the actual dataframe and data2 contains the generated dataframe
accuracy_results = calculate_accuracy(data, data2)

# Display the accuracy results
print("Mode Accuracy: {:.2f}%".format(accuracy_results['Mode Accuracy']))
print("Category Accuracy: {:.2f}%".format(accuracy_results['Category Accuracy']))
print("Overall Accuracy: {:.2f}%".format(accuracy_results['Overall Accuracy']))
```
