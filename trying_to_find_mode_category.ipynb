```python
import pandas as pd
import re

# Sample DataFrame based on the description you provided
data = {
    'Description': [
        'NFS/S1ANMA96/CASH WDL/13-10-19',
        'NFS/S1ANND07/CASH WDL/17-10-19',
        'ATMDEC CHG/13-10-2021/1560+GST',
        'ATM/SPCND037/CASH WDL/27-10-19',
        'NFS/ID102001/CASH WDL/11-11-19',
        'NFS/MC003187/CASH WDL/03-01-20',
        'rem chrgs:atm cw fee-dom/xx6926/723300000000/1...',
        'atm usage charges',
        'atm/debit pin regen chrgs dtd 13/ 151121-mir22...',
        'charge:cash wdl other bank atm/inv291021197257...',
        'limit exceed atm charges in other banks (ref# ...',
        'non scb atm usage charges mar 21'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Function to categorize based on description
def categorize(description):
    # Check for both 'NFS' and 'CASH WDL' in the description
    if re.search(r'nfs', description, re.IGNORECASE) and re.search(r'cash wdl', description, re.IGNORECASE):
        return 'Withdrawal'
    else:
        return 'Charges'

# Apply the function to the 'Description' column
df['Category'] = df['Description'].apply(categorize)

# Show the resulting DataFrame
print(df)
```

```python
import pandas as pd
import re

# Load the CSV file
file_path = '/content/xns.csv'

df = pd.read_csv(file_path)

df
```

```python
df.drop(['Mode', 'Category'], axis=1)
```

```python

def categorize_transaction(description, amount):
    description = description.lower()

    if re.search(r'nfs.*cash wdl', description):
        return 'withdrawal', 'atm'
    elif re.search(r'upi', description):
        return 'other credit', 'upi'
    elif re.search(r'ips', description):
        return 'expenses', 'ips'
    elif re.search(r'atm.*usage.*charge', description):
        return 'charges', 'atm'
    elif re.search(r'bill|rchg|recharge', description):
        return 'expenses', 'bill online'
    elif re.search(r'subsidy|apbs', description):
        return 'subsidy', 'other'
    elif float(amount) < 0:
        return 'expenses', 'other'
    else:
        return 'other credit', 'other'


df[['Category_Predicted', 'Mode_Predicted']] = df.apply(
    lambda row: categorize_transaction(row['Description'], row['Amount']), axis=1, result_type="expand"
)

print(df[['Description', 'Amount', 'Balance', 'Category_Predicted', 'Mode_Predicted']])
```
